```python
def turn_to_int(thing):
    no_dashes = re.sub('\-','',str(thing))[:8]
    return int(no_dashes)

def combine_traf_insta(insta_df, traf_df):
    insta_df['num_posts'] = 1 
    traf_df['date_num'] = traf_df['date'].apply(turn_to_int)
    traf_df.set_index('date_num', inplace=True)
    posts_per_day = insta_df[['num_posts','date_num']].groupby(by='date_num').agg(['count'])
    insta_df_sorted = insta_df.groupby(by=['date','number_of_likes']).max().reset_index()
    insta_df = insta_df[['date','number_of_likes', 'date_num', 'DOW_1', 'DOW_2', 'DOW_3', 'DOW_4', 'DOW_5', 'DOW_6', 'hour','num_people_tagged']].groupby(by=['date_num','number_of_likes']).max().reset_index()    
    best_of_day = insta_df.drop_duplicates(subset=['date'], keep='last')
    best_of_and_count = pd.concat([best_of_day,posts_per_day], axis=1)
    best_of_and_count.set_index('date_num', inplace=True)
    insta_traffic = traf_df[['Instagram', 'total','date']].groupby('date').max()
    new_df = pd.concat([insta_traffic,best_of_and_count],axis=1)
    return new_df
```

```python
df =combine_traf_insta(insta_df,traf_df)
```

```python
def make_best_of_day(insta_df):
    insta_df_sorted = insta_df[['date_num','number_of_likes', 'DOW_1','DOW_2','DOW_3','DOW_4','DOW_5','DOW_6', 'num_people_tagged']].groupby(by=['date_num','number_of_likes']).max().reset_index()
    daily_post_count = insta_df[['date_num','num_posts']].groupby(by=['date_num']).count()
    best_of_day = insta_df_sorted.drop_duplicates(subset=['date_num'], keep='last')
    best_of_day.set_index('date_num', inplace=True)
    best_of_and_count = pd.concat([best_of_day,daily_post_count], axis=1,)
    return best_of_and_count
```

```python
yr2017 = insta_df[insta_df['year']== 2017]
```

```python
gdbr = GradientBoostingRegressor(learning_rate=0.01, loss='ls', max_depth = 2,n_estimators=1000, random_state=1)
```

```python
X = yr2017[['DOW_1','DOW_2','DOW_3','DOW_4','DOW_5','DOW_6','num_people_tagged', 'month', 'hour']]
y = yr2017[['number_of_likes']]
X_train, X_test, y_train, y_test = train_test_split(X.as_matrix(), y.as_matrix(), test_size= .15)
```

```python
gdbr.fit(X_train,y_train)
y_hat=gdbr.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_hat))
```

```python
def bootstrap_resample(X, n=None):
    """ Bootstrap resample an array_like
    Parameters
    ----------
    X : array_like
      data to resample
    n : int, optional
      length of resampled array, equal to len(X) if n==None
    Results
    -------
    returns X_resamples
    """
    if n == None:
        n = len(X)     
    resample_i = np.floor(np.random.rand(n)*len(X)).astype(int)
    X_resample = X[resample_i]
    return X_resample
```

```python
RSEs = []
for i in range(60):
    bootstrap = bootstrap_resample(np.c_[ X_train, y_train ])
    X_train = bootstrap[:, :-1]
    y_train = bootstrap[:, -1]
    gdbr.fit(X_train,y_train)
    y_hat=gdbr.predict(X_test)
    RSEs.append(np.sqrt(mean_squared_error(y_test, y_hat)))
```
